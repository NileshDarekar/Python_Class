### break statement:
 - for breaking the loop prematurely
 - we can also break an infinite loop when some condition is satified

```python
x = 0
while x<7:
    x+=1
    if x==4:
        break
    print(x)

print('out')
```

```python
pressure = 2500

while True:
    print(f'Formation Pressure is {pressure} psi')
    pressure+=100
    
    if pressure ==5000:
        print('Pressure reached its maximum value, so stopping the process')
        break
        
print('Process Finished')
```

```python
pressures = [2300,3200,4200,5000,5200]

for i in pressures:
    print(i)
    
    if i >4000:
        print('breaking the statement')
        break
        
    else:
        print(f'not broken yet, and value of pressure is {i}')
    
```

### Continue : jump back to top[ of loop, rather than stopping it.

 - Stop current iteration and continue with the next one

### Break: Break the loop, no execution of loop at all

```python
for i in  range(7):
    if i ==4:
        print('Skipping 4')
        continue
    print(i)
```

```python
pressure = 3500

while pressure < 4500:
    
    pressure+=100
    
    if pressure ==4000:
        print('Skipping 4000 psi')
        continue
        print('it will never be displayed')
    print(f'Pressure is {pressure} psi')
```

```python
## Sqrt of a number using newton raphson

n = float(input('Hey, Enter your number to be square rooted: '))

x_0 = float(input('Enter your initial guess of the root: '))

x = x_0
count = 0 

while (abs(x**2-n)>0.00000001):
    x = x - ((x**2-n)/(2*x))
    count+=1

print(f"Result: The square root of {n} is {x}")

print(f"The actual square root of {n} is {n**0.5}")

print(f"The num of iterations used = {count}")
```

## Finding Klinkenberg Permeability

```python
# f'(x) = 4.416 x^(-0.36) + pm

#f(x) = 6.9*x^0.64 + pm*x - pm*kg
```

```python
## Kl using newton raphson

kg = float(input('Hey, Enter kg '))
pm = float(input('Hey, Enter pm '))


x_0 = float(input('Enter your initial guess of kl: '))

x = x_0
count = 0 

Fn= 6.9*x**0.64+pm*x-pm*kg
while (abs(Fn)>0.00000001):
    Fn= 6.9*x**0.64+pm*x-pm*kg
    Fnd=4.416*x**(-0.36)+pm
    x = x - ((Fn)/(Fnd))
    count+=1

print(f"Result: Value of Kl is {x}")

print(f"The num of iterations used = {count}")
# f'(x) = 4.416 x^(-0.36) + pm
```
