##### Return : return your datatype as it is, can perform operations

```python
def klinkenberg_perm(kg,pm,x_0):
    """
    This klinkenberg function calculates absolute permeability for given gas permeability at a mean pressure.
    INPUTS:
    kg=> gas permeability measured in lab at a given mean pressure pm
    pm=> mean pressure
    x_0 => User initial guess for absolute permeability
    
    return:
    k_l => Value of absolute permeability
    count => Number of iterations ran
    """

    k_l = x_0
    count = 0 

    Fn= 6.9*k_l**0.64+pm*k_l-pm*kg
    while (abs(Fn)>0.00000001):
        Fn= 6.9*k_l**0.64+pm*k_l-pm*kg
        Fnd=4.416*k_l**(-0.36)+pm
        k_l = k_l - ((Fn)/(Fnd))
        count+=1

    print(f"Result: Value of Kl is {k_l}")

    print(f"The num of iterations used = {count}")
    
    return k_l,count
```

```python
abs_perm,iterations =  klinkenberg_perm(46.6,2.15,20)
```
