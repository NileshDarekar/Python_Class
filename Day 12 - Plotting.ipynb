## Visualizations Libraries:

1. Matplotlib 
2. Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sw  = np.linspace(0,1,100)
```

```python
sw
```

```python
krw = sw**3
```

```python
#krw = Relative permeability of water
krw 
```

```python
#kro = relative permeability of oil 
kro  = (1-sw)**2 
```

```python
kro 
```

```python
df_relperm = pd.DataFrame({'Water Saturation':sw,'Rel Perm Water':krw,'Rel Perm Oil':kro})
```

```python
df_relperm
```

```python
df_relperm.plot()
```

```python
df_relperm.plot(subplots = True)
```

```python
# Step1 : krw vs sw plot
```

```python
plt.plot(sw,krw)
```

```python
# Step2: Kro vs sw plot

plt.plot(sw,kro)
```

```python
# Step 3: Krw and Kro vs sw plot:

plt.plot(sw,krw)
plt.plot(sw,kro)
```

```python
# Step 4: Given the axis labels name:

plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Water Saturation(Fraction)")

plt.ylabel("Relative Permeabilities")
```

```python
# Step 5: Grid

plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Water Saturation(Fraction)")

plt.ylabel("Relative Permeabilities")

plt.grid(True)
```

```python
# Step 6: Adding legend to the plot

plt.plot(sw,krw,label = "Water Rel Perm(Krw)")
plt.plot(sw,kro,label = "Oil Rel Perm(Krw)")

plt.xlabel("Water Saturation(Fraction)")

plt.ylabel("Relative Permeabilities")

plt.grid(True)

plt.legend(loc = 'best')
```

```python
#Step7: Change the size of plot

plt.figure(figsize = (12,6))
plt.plot(sw,krw,label = "Water Rel Perm(Krw)")
plt.plot(sw,kro,label = "Oil Rel Perm(Krw)")

plt.xlabel("Water Saturation(Fraction)")

plt.ylabel("Relative Permeabilities")

plt.grid(True)

plt.legend(loc = 'best')
```

```python
# Step 8: Change the line color

plt.figure(figsize = (12,6))
plt.plot(sw,krw,label = "Water Rel Perm(Krw)",color = 'darkblue')
plt.plot(sw,kro,label = "Oil Rel Perm(Krw)",color = 'brown')

plt.xlabel("Water Saturation(Fraction)")

plt.ylabel("Relative Permeabilities")

plt.grid(True)

plt.legend(loc = 'best')
```

```python
# Step 9: Changing the line thickness
plt.figure(figsize = (12,6))
plt.plot(sw,krw,label = "Water Rel Perm(Krw)",color = 'darkblue',lw = 0.25)
plt.plot(sw,kro,label = "Oil Rel Perm(Krw)",color = 'brown',lw = 4,alpha = 0.2)

plt.xlabel("Water Saturation(Fraction)")

plt.ylabel("Relative Permeabilities")

plt.grid(True)

plt.legend(loc = 'best')
```

```python
# Step 10: title

plt.figure(figsize = (12,6))
plt.plot(sw,krw,label = "Water Rel Perm(Krw)",color = 'darkblue',lw =2)
plt.plot(sw,kro,label = "Oil Rel Perm(Krw)",color = 'brown',lw = 4)

plt.xlabel("Water Saturation(Fraction)")

plt.ylabel("Relative Permeabilities")

plt.grid(True)

plt.legend(loc = 'best')
plt.title('Relative Permeability Plot')
```

```python
# Changing the line styling
plt.figure(figsize = (12,6))
plt.plot(sw,krw,'--',label = "Water Rel Perm(Krw)",color = 'darkblue',lw =2)
plt.plot(sw,kro,'+',label = "Oil Rel Perm(Krw)",color = 'brown',lw = 4)

plt.xlabel("Water Saturation(Fraction)")

plt.ylabel("Relative Permeabilities")

plt.grid(True)

plt.legend(loc = 'best')
plt.title('Relative Permeability Plot')
```
